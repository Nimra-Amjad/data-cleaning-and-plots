```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
plt.rcParams["figure.figsize"] = [10,5]
import warnings
warnings.simplefilter(action="ignore",category=FutureWarning)
```

```python
full_data=pd.read_csv('/content/Titanic-Dataset.csv')
```

```python
print("Train Data: ",full_data.shape)
```

```python
full_data.head(5)
```

```python
full_data.info()
```

```python
sns.heatmap(full_data.isnull(),yticklabels=False,cbar=False,cmap='tab20c_r')
plt.title("Missing Data: Training Set")
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.boxplot(x="Pclass",y="Age",data=full_data,palette="GnBu_d").set_title("Age by passenger class")
plt.show()
```

```python
def inpute_age(cols):
  Age=cols[0]
  Pclass=cols[1]
  if pd.isnull(Age):
    if Pclass==1:
      return 37
    elif Pclass==2:
      return 29
    else:
      return 24
  else:
    return Age

full_data['Age']=full_data[['Age','Pclass']].apply(inpute_age,axis=1)
```

```python
full_data.drop('Cabin',axis=1,inplace=True)
```

```python
full_data.dropna(inplace=True)
```

```python
full_data
```

```python
full_data.drop(['Name','Ticket'],axis=1,inplace=True)
objcat=['Sex','Embarked']
for colname in objcat:
  full_data[colname]=full_data[colname].astype('category')
```

```python
full_data
```

```python
full_data.describe()
```

```python
full_data.drop('PassengerId',axis=1,inplace=True)
```

```python
full_data.shape
```

```python
full_data.select_dtypes(['category']).columns
```

```python
sex=pd.get_dummies(full_data['Sex'],drop_first=True)
embarked=pd.get_dummies(full_data['Embarked'],drop_first=True)
```

```python
full_data.head()
```

```python
full_data=pd.concat([full_data,sex,embarked],axis=1)
full_data.head(5)
```

```python
full_data.drop(['Sex','Embarked'],axis=1,inplace=True)
```
